# Counting Scan Events in a JSON File
This notebook demonstrates how to count the number of scan events in a JSON file containing an array of events, using Python and pandas.

## 1. Import Required Libraries
We will use Python's built-in `json` module for reading the file and `pandas` for convenient data handling.

```python
import json
import pandas as pd
```

## 2. Define the Problem Statement
We want to count how many events in the JSON file have `attributes.type.value` equal to `'scan'`.

## 3. Design the Notebook Workflow
1. Load the JSON file.
2. Extract the list of events.
3. Filter events where `attributes.type.value == 'scan'`.
4. Count and display the result.

## 4. Implement Example Functionality
The following code loads the JSON file, filters for scan events, and counts them.

```python
# Set the path to your JSON file
json_path = '../Downloads/events_1_minute_DFL-MAT-J041YD.json'

# Load the JSON data
with open(json_path, 'r') as f:
    data = json.load(f)

# Extract events
all_events = data['events']

# Filter for scan events
scan_events = [event for event in all_events if event.get('attributes', {}).get('type', {}).get('value') == 'scan']

# Count scan events
num_scan_events = len(scan_events)
print(f"Number of scan events: {num_scan_events}")
```
